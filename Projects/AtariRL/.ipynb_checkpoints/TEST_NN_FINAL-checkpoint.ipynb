```python
import numpy as np
from gym import spaces
from IPython.display import clear_output
from matplotlib import pyplot as plt
```

```python
BLANK_ID = " "
ENEMY_ID = "x"
PLAYER_ID = "p"
WALL_ID = "#"
END_ID = "o"

class Environment:
    def __init__(self, name: str = "CustomEnvironment"):
        self.name = name
        self.mapsize = (8,8) # scalable in future
        self.nrows = self.mapsize[0]
        self.ncols = self.mapsize[1]
        self.endposition = (6,6)
        self.action_space = ['up','down','left','right']
        self.action_size = len(self.action_space)
        self.state_size = self.nrows * self.ncols
        self.state_space = spaces.Discrete(self.state_size)
        self.map = self._generate_random_map(5)
        
        
        self.oldstate = None
        self.state = None
    def _generate_random_map(self, nenemies: int):
        envmap = np.full(self.mapsize, BLANK_ID)
        envmap[:,[0,-1]] = envmap[[0,-1]] = WALL_ID
        envmap[self.endposition] = END_ID
        for i in range(nenemies):
            validpos = False
            x = 0
            y = 0
            while not validpos:
                x = np.random.randint(0, self.nrows)
                y = np.random.randint(0, self.ncols)
                validpos = envmap[x][y] == BLANK_ID and (x,y) != self.endposition
            envmap[x][y] = ENEMY_ID
        return envmap
    
    def is_wall(self, x, y):
        return self.map[x][y] == WALL_ID

    def is_enemy(self, x, y):
        return self.map[x][y] == ENEMY_ID
    
    def is_end(self, x, y):
        return self.map[x][y] == END_ID
    def is_terminal(self, x, y):
        return self.is_wall(x,y) or self.is_end(x,y)
    
    def show_map(self):
        if self.map is not None:
            print(self.map)
    
    def to_state(self, *args):
        if len(args)==1:
            assert isinstance(args[0], tuple)
            state = args[0]
            assert len(state)==2, f"[ToState] Tuple state passed does not have size 2, but {len(state)}"
            assert isinstance(state[0], int) and isinstance(state[1], int), "[ToState] Elements of state must be int."
            # (x,y) passed
            x,y = state
            return x * self.ncols + y
        elif len(args)==2:
            x = args[0]
            y = args[1]
            assert isinstance(x, int) and isinstance(y, int), "[ToState] x, y values must be integers"
            return x * self.ncols + y
        else:
            raise ValueError(f"[ToState] Invalid number of argumnets passed: {len(args)}")
    
    
    def step(self, action):
        oldx, oldy = self.state
        self.oldstate = (oldx, oldy)
        x,y = self.move(oldx, oldy, action)
        if self.is_wall(x,y):
            reward = -100
        elif self.is_enemy(x,y):
            reward = -60
        elif self.is_end(x,y):
            reward = 200
        else:
            reward = -1
        done = self.is_wall(x,y) or self.is_end(x,y)
        self.state = (x,y)
        return (x,y),reward, done, False # False is useless, just to match gym returns
        
    def render(self):
        clear_output(wait=True)
        newmap = self.map
        newmap[self.state] = PLAYER_ID
        if self.oldstate is not None:
            newmap[self.oldstate] = BLANK_ID
        print(newmap)
    
    def reset(self):
        validpos = False
        x = 0
        y = 0
        while not validpos:
            x = np.random.randint(0, self.nrows)
            y = np.random.randint(0, self.ncols)
            validpos = (not self.is_end(x,y)) and (not self.is_wall(x,y)) and (not self.is_enemy(x,y))
        self.state = (x,y)
        return self.state
    
    def move(self, x, y, actionindex):
        if self.action_space[actionindex] == "up":
            x = max(x - 1, 0)
        elif self.action_space[actionindex] == "left":
            y = max(y - 1, 0)
        elif self.action_space[actionindex] == "down":
            x = min(x + 1, self.nrows - 1)
        elif self.action_space[actionindex] == "right":
            y = min(y + 1, self.ncols - 1)
        return (x, y)
            
    
        
        
```

```python
e = Environment()


LEARNING_RATE = 0.1
DISCOUNT = 0.95
EPISODES = 10000
STATS_EVERY = 100
max_epoch_iterations = 50000
epsilon = 1  # not a constant, qoing to be decayed
START_EPSILON_DECAYING = 1
END_EPSILON_DECAYING = EPISODES
epsilon_decay_value = epsilon/(END_EPSILON_DECAYING - START_EPSILON_DECAYING)

q_table = np.random.uniform(low=-2, high=0, size=(e.state_size, e.action_size))
print(f"Initialized q-table with shape {q_table.shape}")

episode_iterations = []
episode_rewards = []
aggr_ep_rewards = {'ep': [], 'avg': [], 'max': [], 'min': []}

for episode in range(EPISODES):
    print(F"Starting episode {episode+1}/{EPISODES}", end='\r', flush=True)
    startingstate = e.reset()
    state = e.to_state(startingstate)
    episode_reward = 0
    ep_iter = 0
    done = False
    for time in range(max_epoch_iterations):
        #print("Time: ",time)
        if np.random.random() > epsilon:
            # Get action from Q table
            action = np.argmax(q_table[state])
        else:
            # Get random action
            action = np.random.randint(0, e.action_size)
        new_state, reward, done, _ = e.step(action_index)
        episode_reward += reward
        
        new_state_val = e.to_state(new_state)
        if not done:
            # Maximum possible Q value in next step (for new state)
            max_future_q = np.max(q_table[new_state_val])
            # Current Q value (for current state and performed action)
            current_q = q_table[state][action_index]
            # And here's our equation for a new Q value for current state and action
            new_q = (1 - LEARNING_RATE) * current_q + LEARNING_RATE * (reward + DISCOUNT * max_future_q)
            # Update Q table with new Q value
            q_table[state][action_index] = new_q
        else:
            q_table[state][action] = 0
            #print("Start: ",startingstate," - END: ",new_state, " - iter: ",ep_iter)
            break
        ep_iter += 1
        state = new_state_val
        if END_EPSILON_DECAYING >= episode >= START_EPSILON_DECAYING:
            epsilon -= epsilon_decay_value
    
    episode_rewards.append(episode_reward)
    if not episode % STATS_EVERY:
        average_reward = sum(episode_rewards[-STATS_EVERY:])/STATS_EVERY
        aggr_ep_rewards['ep'].append(episode)
        aggr_ep_rewards['avg'].append(average_reward)
        aggr_ep_rewards['max'].append(max(episode_rewards[-STATS_EVERY:]))
        aggr_ep_rewards['min'].append(min(episode_rewards[-STATS_EVERY:]))
        print(f'Episode: {episode:>5d}, average reward: {average_reward:>4.1f}, current epsilon: {epsilon:>1.2f}')
    episode_iterations.append(ep_iter)
    #episode_rewards.append(ep_reward)
```

```python
plt.plot(aggr_ep_rewards['ep'], aggr_ep_rewards['avg'], label="average rewards")
plt.plot(aggr_ep_rewards['ep'], aggr_ep_rewards['max'], label="max rewards")
plt.plot(aggr_ep_rewards['ep'], aggr_ep_rewards['min'], label="min rewards")
plt.legend(loc=4)
plt.show()
```
