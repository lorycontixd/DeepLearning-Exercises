```python
import os
import seaborn as sn
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

dataset_folder = "dataset/"
```

```python
df = pd.read_csv(os.path.join(dataset_folder, "ds_salaries.csv"))
print(list(df.columns))
y = df['salary_in_usd']
x = df.copy()
# Only keep salary in usd (so only 1 conversion is needed)
x.drop(['Unnamed: 0', 'salary', 'salary_currency', 'salary_in_usd'], axis=1, inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(14,7))
plt.scatter(x['experience_level'],y)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(14,7))
plt.scatter(x['job_title'],y)
plt.xticks(rotation='vertical')
plt.show()
```

```python
salary_means= df.groupby('job_title', as_index=False)['salary_in_usd'].mean()
fig, ax = plt.subplots(figsize=(14,7))
plt.bar(salary_means['job_title'],salary_means['salary_in_usd'])
plt.xticks(rotation='vertical')
plt.show()
```
